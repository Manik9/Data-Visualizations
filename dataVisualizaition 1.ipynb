```python
import matplotlib.pyplot as plt
x = np.arange(1, 10)
y = x.reshape(-1, 1)
h = x * y

cs = plt.contourf(h, levels=[10, 30, 50],
    colors=['#808080', '#A0A0A0', '#C0C0C0'], extend='both')
cs.cmap.set_over('red')
cs.cmap.set_under('blue')
cs.changed()
```

```python
X=np.random.randn(100)
```

```python
fig,ax=plt.subplots(5,figsize=(5,20),dpi=40,ncols=2)
for i in range(10):
    X=np.random.randn(10**i)
    if i%2:
        ax[i//2][0].hist(X)
    else:
        ax[i//2][1].hist(X,log=True)
```

```python
fig,ax=plt.subplots((7),figsize=(5,20),dpi=40)
for i in range(7):
    X=np.random.randn(10**i)
    ax[i].hist(X,log=True)
```

```python
x=np.linspace(0,2,100)
fig,ax=plt.subplots(5,figsize=(5,20),dpi=50)
ax[0].plot(x,x**1,color='red')
ax[1].plot(x,x**2,color='green')
ax[2].plot(x,x**3,color='blue')
ax[3].plot(x,x**4,color='pink')
ax[4].plot(x,x**5,'purple')
```
